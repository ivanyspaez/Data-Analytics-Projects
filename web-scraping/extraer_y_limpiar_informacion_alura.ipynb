```python
import requests
from bs4 import BeautifulSoup
```

```python

# Función para extraer el contenido detallado de cada curso
def extract_detailed_content(url):
    try:
        response = requests.get(url, allow_redirects=True, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar el título de "Contenido Detallado"
        detailed_content_title = soup.find('h2', class_='courseInfo-detailed-content-title')
        if not detailed_content_title:
            return None

        # Extraer todo el contenido hasta el último div antes del enlace CSS
        detailed_content = []
        for tag in detailed_content_title.find_all_next():
            if tag.name == 'link' and 'block/planosPagamentoOld.1720710819.css' in tag.get('href', ''):
                break
            detailed_content.append(str(tag))

        return ''.join(detailed_content)

    except requests.exceptions.RequestException as e:
        print(f"Error al procesar {url}: {e}")
        return None

# Lista para almacenar el contenido detallado
detailed_contents = []
problematic_urls = []

# Recorrer la lista de cursos y extraer el contenido detallado
for course in course_info_list:
    print(f"Extrayendo contenido detallado de: {course['title']}")
    content = extract_detailed_content(course['url'])
    if content:
        detailed_contents.append({
            "title": course['title'],
            "url": course['url'],
            "content": content
        })
    else:
        detailed_contents.append({
            "title": course['title'],
            "url": course['url'],
            "content": "No se encontró contenido detallado."
        })
        problematic_urls.append(course['url'])

# Mostrar el contenido extraído
for detail in detailed_contents:
    print(f"Título: {detail['title']}")
    print(f"URL: {detail['url']}")
    print(f"Contenido: {detail['content']}")
    print("\n" + "="*80 + "\n")

# Mostrar URLs problemáticas
if problematic_urls:
    print("URLs problemáticas:")
    for url in problematic_urls:
        print(url)

# O puedes guardar la lista detailed_contents para procesarla más adelante
```

```python
def guardar_cursos_en_texto(cursos, archivo):
    with open(archivo, 'w', encoding='utf-8') as file:
        for curso in cursos:
            file.write(f"Curso: {curso['title']}\n")
            file.write(f"Link: {curso['url']}\n")
            file.write("Contenido Detallado:\n")
            file.write(f"{curso['content']}\n")
            file.write("=" * 40 + "\n\n")

archivo = "cursos.txt"
guardar_cursos_en_texto(cursos, archivo)
print(f"La información se ha guardado en {archivo}")
print("para obtener los resultados, descargue el archivo.txt")
```
